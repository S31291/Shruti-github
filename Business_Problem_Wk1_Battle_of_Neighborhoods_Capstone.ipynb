Week 1: Battle of the Neighborhoods- IBM - Applied Data Science Capstone project 

Business Problem : To help a customer decide where to open a restaurant in Bangalore.

Introduction: A customer wants to open a restaurant in Bangalore. He is looking for a potential neigborhood in bangalore having 
many companies with good employee size, to open his retaurant.

Methodology used to solve the problem:

1.Data collection and data sources

i. Dataset 1 -Pre-Processing of Neighborhood data of Bangalore\
Neighborhood data of Bangalore with corresponding pincodes and geograhical data (longitudes and latittudes) was pre-processed into a .csv file from folder IN.zip.\
This particulaar folder was obtained from the GeoNames geographical database-a reliable database for geographical and pincode data of listed countries.\
The datasets are free for downloaing under the  Creative Commons - Attribution 3.0 Unported - CC BY 3.0 licence.

ii. Dataset 2- Data (number) of all companies in Bangalore, segregated according to the number of employees and pincodes\
This particular data set was obtained from a wesite:' https://www.fundoodata.com/products/advance_search.php?wt=&rws=1 '.This website allows the user to find the 
number of companies in a particular neighbourhood (pincode) of Bangalore. Addtionlly, the user can find the number of companies located within a neigborhood with
a specific employee size. Employee size categories provided by the wesite are: 1-50,51-100,101-250,251-500,501-1000,1001-2500,2501-5000 and 5001 & above.

iii. Data set 3- Foursquare data of all types of food outlets (search_queries- cafe, bar, restaurant, bistro) located within 500m of the longitudes and latitude of the neighborhoods of Bangalore.\
The data was retrieved after creating a developer account in Foursqaure website :' https://foursquare.com/ ' followed by using generated client ID and client secret. 

2.Integrating the three datasets into a pandas dataframe.

3.Clustering the neighborhhods of Bangalore according to their geographical data using K-means clustering algorithm in Python.

4.Identifying the cluster with Banglore neighborhoods having both the highest total employee size (depends on number of companies and their employee sizes) and lowest number of food outlets.

5.Suggesting the identified neighborhood to the cuatomer as a potential area to open a restaurant.
