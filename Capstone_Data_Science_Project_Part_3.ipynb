Removing all rows from dataframe with Boroughs designated as 'Not Assigned'.

```python
dffinal=new_df.reset_index(drop=True)
dffinal
```

Grouping the Neigborhoods by Postal Code and Borough and then reseting the index of the dataframe.

```python
dffinal = dffinal.groupby(["Postal Code", "Borough"])["Neighborhood"].apply(", ".join).reset_index()
dffinal
```

Part 2. Retrieving the geographical co-ordinates from the link: http://cocl.us/Geospatial_data and merging it with generated dataframe.

```python
dfll=pd.read_csv("http://cocl.us/Geospatial_data")
dfll
```

```python
df_toronto = pd.merge(dffinal, dfll, how='left', left_on = 'Postal Code', right_on = 'Postal Code')
df_toronto
```

Part 3. Clustering Neighborhoods according to geographical co-ordinates mentioned in generted dataframe, with Boroughs ending with 'Toronto'.

Importing all required libraries.

```python
import numpy as np
import json
from geopy.geocoders import Nominatim 
import matplotlib.cm as cm
import matplotlib.colors as colors
from sklearn.cluster import KMeans
!conda install -c conda-forge folium=0.5.0 --ye
import folium
print('Libraries imported.')
```

Selecting rows with 'Boroughs' containing string 'Toronto' and saving it in a new dataframe called new_df_toronto.

```python
new_df_toronto=df_toronto[df_toronto.Borough.str.contains('Toronto', case=False)]
new_df_toronto=new_df_toronto.reset_index(drop=True)
new_df_toronto
```

Printing the size of new_df_toronto.

```python
new_df_toronto.shape
```

Genertaing a map of Boroughs in Toronto with their geographical co-ordinates, in new_df_toronto using Folium library.

```python
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(new_df_toronto['Latitude'],new_df_toronto['Longitude'],new_df_toronto['Borough'],new_df_toronto['Neighborhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

Clustering the Neighbourhoods in Toronto using K-means clustering

```python
k=5
toronto_clustering = new_df_toronto.drop(['Postal Code','Borough','Neighborhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
new_df_toronto.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
new_df_toronto
```

Genertaing map of the clustered Neighborhoods in Toronto after application of K-means Clustering to dataframe new_df_toronto.

```python
# create map
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(new_df_toronto['Latitude'], new_df_toronto['Longitude'], new_df_toronto['Neighborhood'], new_df_toronto['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
